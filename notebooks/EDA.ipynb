## Verificacion de datos coherentes

```python
def verificar_fechas(df, col_fecha='fecha', col_anio='anio', col_mes='mes', col_hora='hora', col_dia='dia_semana'):
    """
    Verifica que las columnas de fecha, año, mes y hora sean coherentes.
    - anio=0 debe ser 2011, anio=1 debe ser 2012.
    - Devuelve un DataFrame solo con las filas inconsistentes, sin modificar el original.
    """
    # Hacer una copia para no modificar el original
    temp_df = df.copy()
    
    # Convertir a datetime y extraer componentes
    temp_df[col_fecha] = pd.to_datetime(temp_df[col_fecha])
    fecha_anio = temp_df[col_fecha].dt.year
    fecha_mes = temp_df[col_fecha].dt.month
    fecha_dia = ((temp_df[col_fecha].dt.dayofweek)+1)%7
    
    # Mapear anio numérico a año real
    anio_real = temp_df[col_anio].apply(lambda x: 2011 if x == 0 else (2012 if x == 1 else None))
    
    # Verificar inconsistencias
    mask = (
        (fecha_anio != anio_real) | 
        (fecha_mes != temp_df[col_mes]) |
        (fecha_dia != temp_df[col_dia])
        #(temp_df[col_hora] != temp_df[col_fecha].dt.hour)
    )
    
    # Filtrar solo filas inconsistentes
    inconsistentes = temp_df[mask].copy()
    
    # Agregar columnas temporales solo al resultado
    inconsistentes['fecha_anio'] = fecha_anio[mask]
    inconsistentes['fecha_mes'] = fecha_mes[mask]
    inconsistentes['fecha_dia'] = fecha_dia[mask]
    inconsistentes['anio_real'] = anio_real[mask]
    
    return inconsistentes[[col_fecha, col_anio, col_mes, col_dia, col_hora, 'fecha_anio', 'fecha_mes', 'fecha_dia', 'anio_real']]

# Uso ejemplo
inconsistentes = verificar_fechas(df)
len(inconsistentes)
inconsistentes
```

mes, dia_semana

## Análisis de Outliers

```python
df.describe()
```

Se observa que el 75% de los datos tanto de sensacion_termica como de velocidad_viento son menores a 1, pero sus maximos son de 803 y 806 
respectivamente. Procedemos a verificar el rango intercuartilico. 

```python
Q1 = df['sensacion_termica'].quantile(0.25)
Q3 = df['sensacion_termica'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
outliers = (df['sensacion_termica'] < lower_bound) | (df['sensacion_termica'] > upper_bound)

# Contar la cantidad de outliers
num_outliers = outliers.sum()

print(f"Número de outliers: {num_outliers}")
print(f"Porcentaje de outliers: {num_outliers/len(df)*100:.2f}%")
```

```python
Q1 = df['velocidad_viento'].quantile(0.25)
Q3 = df['velocidad_viento'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
outliers = (df['velocidad_viento'] < lower_bound) | (df['velocidad_viento'] > upper_bound)

# Contar la cantidad de outliers
num_outliers = outliers.sum()

print(f"Número de outliers: {num_outliers}")
print(f"Porcentaje de outliers: {num_outliers/len(df)*100:.2f}%")
```

## Transformaciones

Las variables numericas ya estan normalizadas, pero usaremos Standard Scaler y para las categoricas one hot encoding. 

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
if 'fecha' in df.columns:
    df.drop(columns=['fecha'], inplace=True)
```

```python
# Identificar columnas numéricas y categóricas
numeric_features = ['temperatura', 'humedad', 'velocidad_viento']
categorical_features = ['temporada', 'anio', 'mes', 'hora', 'feriado', 'dia_semana', 'dia_trabajo', 'clima']

# Crear el ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

```

```python
# Ajustar y transformar
X_transformed = preprocessor.fit_transform(df)

# Obtener nombres de columnas transformadas
num_features_names = numeric_features
cat_features_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)

# Unir nombres de columnas
all_feature_names = list(num_features_names) + list(cat_features_names)

# Crear DataFrame transformado
df_transformed = pd.DataFrame(X_transformed.toarray() if hasattr(X_transformed, 'toarray') else X_transformed,
                              columns=all_feature_names)

df_transformed.head()
```

De la misma observamos como queda la distribucion de la variable target total_alquileres despues de aplicar una transformacion log

```python
y = df['total_alquileres']
y_log = np.log1p(y)

plt.figure(figsize=(10, 6))
sns.histplot(y, bins=50, kde=True)
plt.title('Distribución de total_alquileres')
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(y_log, bins=50, kde=True)
plt.title('Distribución de log(total_alquileres + 1)')
plt.show()
```

Observamos que despues de realizar la trasformacion el sesgo disminuye lo cual puede beneficiar a nuestros modelos de regresión 
ya que estos trabajan mejor cuando la distribucion se parece más a una distribucion normal
